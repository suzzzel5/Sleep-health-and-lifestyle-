```python
# Analyze the quality of sleep 
#histrogram
sns.histplot(dataset["Quality of Sleep"])
plt.savefig("../Figures/Quality_of_sleep.png")
plt.show()
```

```python
# Checking the sleep duration 

#histogram
sns.histplot(dataset['Sleep Duration'])
plt.savefig("../Figures/Sleep duration ")
plt.show()
```

```python
numerical_cols = ['Age', 'Physical Activity Level', 
                    'Stress Level', 'Heart Rate', 'Daily Steps']

fig, axes = plt.subplots(len(numerical_cols), 2, figsize=(12, 20))

for i, col in enumerate(numerical_cols):
     sns.histplot(dataset[col], ax=axes[i, 0], color='steelblue')
     axes[i, 0].set_title(f'{col} — Histogram')

     sns.boxplot(y=dataset[col], ax=axes[i, 1], color='steelblue')
     axes[i, 1].set_title(f'{col} — Boxplot')

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
     dataset[col].value_counts().plot(kind='bar', ax=axes[i], color='mediumpurple')
     axes[i].set_title(f'{col} — Count')
     axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
